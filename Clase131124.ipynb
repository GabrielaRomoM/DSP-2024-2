```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t1 = np.linspace(0, 5, 100)  # 100 samples
t2 = np.linspace(0, 5, 800)  # 500 samples
f = 1600
s1 = 10 * np.sin(2 * np.pi * f * t1)
s2 = 10 * np.sin(2 * np.pi * f * t2)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(t1, s1, label='Señal 1')
plt.plot(t2, s2, label='Señal 2')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate FFT for both signals
fft_s1 = np.fft.fft(s1)
fft_s2 = np.fft.fft(s2)

# Calculate frequencies for both signals
freq_s1 = np.fft.fftfreq(len(s1), d=t1[1]-t1[0])
freq_s2 = np.fft.fftfreq(len(s2), d=t2[1]-t2[0])

# Create the plots
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(freq_s1, np.abs(fft_s1))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('100 puntos')
plt.xlim(-10, 10)
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(freq_s2, np.abs(fft_s2))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('1000 puntos')
plt.xlim(-10, 10)
plt.grid(True)

plt.tight_layout()
plt.show()
```
