Taller 1:

1.   Generar señal con sample rate >> (grande).
2.   Generar señal con sample rate << (pequeño).
3.   Las dos señales en un interavalo de tiempo (0,1).
4.   Calcular FFT para ambas señales, gráficar y comparar.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
f = 60  # Frequency of the sine waves in Hz
t = 1   # Duration in seconds
fs1 = 3000
fs2 = 500

# Time arrays
t1 = np.linspace(0, t, fs1)  # 3000 samples per second
t2 = np.linspace(0, t, fs2)  # 500 samples per second

# Generate the sine waves
s1 = np.sin(2 * np.pi * f * t1)
s2 = np.sin(2 * np.pi * f * t2)

# Plot the signals
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(t1, s1)
plt.title("3000 samples")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.subplot(1, 2, 2)
plt.plot(t2, s2)
plt.title("500 samples")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#FFT
fft1 = np.fft.fft(s1)
freq1 = np.fft.fftfreq(len(fft1), 1 / fs1)

fft2 = np.fft.fft(s2)
freq2 = np.fft.fftfreq(len(fft2), 1 / fs2)

# Plot the signals
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(freq1[:len(freq1)//2], np.abs(fft1[:len(fft1)//2]), label='FFT 1 (Fs = 3000 Hz)')
plt.title("3000 samples")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.xlim(0,100)
plt.grid(True)
plt.subplot(1, 2, 2)
plt.plot(freq2[:len(freq2)//2], np.abs(fft2[:len(fft2)//2]), label='FFT 2 (Fs = 500 Hz)', color='orange')
plt.title("500 samples")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.xlim(0,100)
plt.grid(True)
plt.tight_layout()
plt.show()
```
