```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output
!pip install PyWavelets
import pywt
```

# Tarea 3 (Parte 1):

```python
slider = widgets.FloatSlider(value=60, min=0, max=100, step=10, description="Frecuencia:")

def grafica(f):
    t = np.linspace(0, 1, 1000)
    x = np.sin(2 * np.pi * 5 * t) + np.sin(2 * np.pi * f * t)
    x2 = np.sin(2 * np.pi * f * t)
    convol = np.convolve(x, x2)

    plt.figure(figsize=(14, 2))
    plt.subplot(121)
    plt.plot(x, label="x = sin(2π5t) + sin(2πft)") ; plt.title("Señal x(t)")
    plt.subplot(122)
    plt.plot(x2, label="x2 = sin(2πft)", color='g') ; plt.title("Señal x2(t)")
    plt.show()
    plt.figure(figsize=(13, 2))
    plt.plot(convol, label="Convolución", color='r') ; plt.title("Convolución de x y x2")
    plt.show()

#Gráfico Interactivo
widgets.interactive(grafica, f= slider)
```

## Convolución con Wavelets

```python
x = np.sin(np.linspace(0, 20* np.pi, 100)) + np.random.normal(0, 0.1, 100)
x2 = np.cos(np.linspace(0, 20* np.pi, 100)) + np.random.normal(0, 0.1, 100)

w = 'db3'
a_x, d_x = pywt.dwt(x, w) ; a_x2, d_x2 = pywt.dwt(x2, w)
convol = np.concatenate((a_x, d_x))

plt.figure(figsize=(12,5))
plt.plot(convol, label='Coeficientes Wavelet')
plt.legend(); plt.show()
```

```python
slider = widgets.FloatSlider(value=60, min=0, max=100, step=10, description="Frecuencia:")

def grafica(f):
    t = np.linspace(0, 1, 1000)
    x = np.sin(2 * np.pi * 5 * t) + np.sin(2 * np.pi * f * t)
    x2 = np.sin(2 * np.pi * f * t)
    x2n = np.copy(x2); x2n[300:] = 0
    convol = np.convolve(x, x2n)

    plt.figure(figsize=(14, 2))
    plt.subplot(121)
    plt.plot(x, label="x = sin(2π5t) + sin(2πft)"); plt.title("Señal x(t)")
    plt.subplot(122)
    plt.plot(x2n, label="x2 = sin(2πft)", color='g')
    plt.title("Señal x2(t)"); plt.show()
    plt.figure(figsize=(13, 2))
    plt.plot(convol, label="Convolución", color='r')
    plt.title("Convolución de x y x2"); plt.show()

#Gráfico Interactivo
widgets.interactive(grafica, f=slider)
```
